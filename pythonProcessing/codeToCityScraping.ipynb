```python
import requests
import pandas as pd
import re
import os
```

```python
airport_code_html = None
tables = None
with open(os.path.join(os.getcwd(), "airportCodes.html"), "r") as my_file:
    airport_code_html = my_file


    # result = requests.get(AIRPORT_CODE_URL)
    # print(result.text)
    tables = pd.read_html(airport_code_html)
    print(len(tables))
```

```python
updated_tables = []
for table in tables:
    table = table.drop([0,1])
    updated_tables.append(table)
```

```python
for table in updated_tables:
    print(table)
```

```python
country_code_info = pd.concat(updated_tables)
country_code_info.columns =['City', 'Country', 'AirportCode']
country_code_info = country_code_info[(country_code_info.Country == "USA")]
country_code_info.set_index('AirportCode', inplace=True)
for index, row_data in country_code_info.iterrows():
    result = re.search(r"[a-zA-Z\s]+", row_data['City']).group()[:-1]
    country_code_info.at[index, 'City'] = result
print(country_code_info)
```

```python
country_code_info.to_csv(os.path.join(os.getcwd(), "airportCodes.csv"))
```
